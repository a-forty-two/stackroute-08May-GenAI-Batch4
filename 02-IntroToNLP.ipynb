```python
mytext = ["Today is a good day but i am unhappy", "the cat named Peanut is overjoyed", "the hungry dog called Rogers is sleepy", "I have a kangaroo who loves boxing and acting like a racoon", 'Today is a good day but i am unhappy']
for text in mytext:
    toks = nlp(text)
    print(text, end = " ; sentiment = ")
    print(toks.sentiment)
```

```python
for text in mytext:
    toks = nlp(text)
    for tok in toks:
        print(tok, end=' pos = ')
        print(tok.pos_)
```

```python
for text in mytext:
    toks = nlp(text)
    print(toks.ents)
    print()
    for tok in toks.ents:
        print(tok)
```

```python
from spacy import displacy
doc = nlp(mytext[1])
displacy.render(doc, style='dep', jupyter=True)
```
